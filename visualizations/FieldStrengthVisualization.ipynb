```python
import json
f = open('data4.json')
data = json.loads(f.readline())
data.reverse()
rows = []
for i in range(len(data)//18):
    rows.append(data[i*18:(i+1)*18])
    rows[i].reverse()
data = [item for row in rows for item in row]
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_raw = np.array([point['coord'][0] for point in data])
y_raw = np.array([point['coord'][1] for point in data])
z = np.array([point['x'] for point in data])
v = np.array([point['y'] for point in data])
b = np.array([point['z'] for point in data])

x=np.unique(x_raw)
y=np.unique(y_raw)
X,Y = np.meshgrid(x,y)

rawZ=z.reshape(len(y),len(x))
rawV=v.reshape(len(y),len(x))
rawB=b.reshape(len(y),len(x))


plt.pcolormesh(X,Y,rawZ)
plt.show()
plt.pcolormesh(X,Y,rawV)
plt.show()
plt.pcolormesh(X,Y,rawB)
plt.show()

#Raw B-field magnitude data (with earth's magnetic field filtered out) in X, Y, Z directions respectively
#(in the frame of the device) for cartesian coordinates of magnet on 2d plane
```

```python
Zc=np.cbrt(rawZ)
Vc=np.cbrt(rawV)
Bc=np.cbrt(rawB)

plt.pcolormesh(X,Y,Zc)
plt.show()
plt.pcolormesh(X,Y,Vc)
plt.show()
plt.pcolormesh(X,Y,Bc)
plt.show()

#Cubed root of the magnitude data
```

```python
Z=np.reciprocal(Zc)
V=np.reciprocal(Vc)
B=np.reciprocal(Bc)

plt.pcolormesh(X,Y,Z)
plt.show()
plt.pcolormesh(X,Y,V)
plt.show()
plt.pcolormesh(X,Y,B)
plt.show()

#Reciprocal of the cubed root of the magnitudes
```

```python
magnitude = np.cbrt(np.sqrt(np.square(rawZ) + np.square(rawV) +np.square(rawB)))
plt.pcolormesh(X,Y,magnitude)
plt.show()

#Total magnitude of magnetic field strength by cartesian coordinates of the magnet
```
